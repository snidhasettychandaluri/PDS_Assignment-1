```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
frail_clean=pd.read_csv('clean_frailty.csv')
```

```python
frail_clean
```

```python
plt.figure(figsize=(8, 6))
plt.hist(frail_clean['Height (Inches)'], bins=10, color='orange', edgecolor='black')
plt.title('Height Distribution')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(frail_clean['Age'], frail_clean['Grip strength'], color='red')
plt.title('Grip Strength vs. Age')
plt.xlabel('Age')
plt.ylabel('Grip Strength')
plt.grid(True)
plt.show()
```

```python
frailty_counts = frail_clean['Frailty'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(frailty_counts.index, frailty_counts.values, color=['orange', 'black'])
plt.title('Frailty Status')
plt.xlabel('Frailty')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```
