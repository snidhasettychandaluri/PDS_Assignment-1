```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
performance=pd.read_csv('StudentsPerformance_clean.csv')
```

```python
performance
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=performance, x='math score', hue='gender', kde=True)
plt.title('Histogram of Math Scores by Gender')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.legend(title='Gender', labels=['Female', 'Male'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=performance, x='race/ethnicity', y='Finals')
plt.title('Boxplot of Reading Scores by Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Reading Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=performance, x='lunch')
plt.title('Bar Chart of Lunch Types')
plt.xlabel('Lunch Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=performance, x='math score', y='Finals')
plt.title('Scatter Plot of Writing Scores vs. Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(performance['math score'],performance['Finals'],marker='o', linestyle='-')
plt.title('Line Plot of Math Scores')
plt.xlabel('math Index')
plt.ylabel('final Score')
plt.show()
```
